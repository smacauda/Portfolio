### Word count

```python
def word_count(str):
    counts = dict()
    words = str.split()
    
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
            
    return counts

def tweet_counts(tweet_array, threshold=10):
    unique, counts = np.unique(np.array(tweet_array), return_counts=True)
    count_dict = dict(zip(unique, counts))
    
    return [(k,v) for k, v in count_dict.items() if v > threshold]
```

```python
who_they_reply_to = []
for tweet in tweets['text']:
    if (tweet[:1]=='@'):
        parts = tweet.split()
        for i in range(0, len(parts)):
            if (parts[i][0]=='@'):
                who_they_reply_to.append(parts[i])
```

```python
tweet_counts(who_they_reply_to, 5)
```
